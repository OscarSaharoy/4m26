```python
A = [1,4,6,3,2,5,7,8,9,6,4,2,4,7,5,4,3,2,3]

def left_child( x ):
    return 2*x + 1

def right_child( x ):
    return 2*x + 2

def swap( A, i, j ):
    A[i], A[j] = A[j], A[i]

def max_heapify( A, source, upto=None ):
    
    upto = upto or len(A)
    biggest = max( [source, left_child(source), right_child(source)],
                   key=lambda x: A[x] if x < upto else -1e9 )
    
    if biggest is source:
        return

    swap( A, source, biggest )
    max_heapify( A, biggest, upto=upto )

        
def heapsort( A ):
    
    for i in range( len(A) // 2, -1, -1 ):
        max_heapify( A, i )
        
    for i, _ in enumerate(A):
        max_heapify( A, 0, upto=len(A) - i )
        swap( A, 0, len(A) - 1 - i )
        
    return A

heapsort(A)
```

```python
B = [64,57,96,35,678,53,24,78,5,23,57,97,54,22,45,7,967,84,73,37,79]

def dist( a, b ):
    return abs( a - b )
    #return sum([ (c1 - c2)**2 for c1, c2 in zip(a,b) ]) ** .5


def closest( A ):
    
    if len(A) < 2:
        return -1e9, 1e9
    
    if len(A) == 2:
        return A[0], A[1]
    
    pivot = sum(A)/len(A)
    
    same  = [ x for x in A if x == pivot ]
    if len(same) > 1:
        return same[:2]
    
    left  = [ x for x in A if x <= pivot ]
    right = [ x for x in A if x >  pivot ]
    
    closest_left = closest(left)
    closest_right = closest(right)
    delta = min( dist(*closest_left), dist(*closest_right) )
    
    within = [ x for x in A if abs( x - pivot ) < delta ]
    
    closest_within = closest(within)
    
    return min( closest_left, closest_right, closest_within, key=lambda t: dist(*t) )


def closest2( A ):
    
    S = sorted(A)
    T = [ (a, b, abs(a-b)) for a,b in zip(S[:-1], S[1:]) ]
    
    return min( T, key = lambda x: x[2] )[:2]


def closest3( A ):
    
    pairs = [ (a,b,abs(a-b)) for i,a in enumerate(A) for b in A[i+1:] ]
    
    return min( pairs, key = lambda x: x[2] )[:2]
```

```python
A = [64,57,96,35,678,53,24,78,5,23,57,97,54,22,45,7,67,84,73,37,79]

def two_sum( A, target ):
    
    seen = set()
    
    for a in A:
        if target - a in seen:
            return a, target - a
        seen.add( a )
        
        
def three_sum( A, target ):
    
    for a in A:
        if t := two_sum( A, target - a ):
            return a, *t
    
        
three_sum( A, 100 )
```

```python
from IPython.core.display import display, HTML
display(HTML("<style>div.output_area pre {white-space: pre;}</style>"))


class Node:
    def __init__(self, key):
        self.key = key
        self.right = None
        self.left = None

class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, key):
        if not self.root:
            self.root = Node(key)
        
        node = self.root
        
        while True:
            
            if node.key <= key:
                if not node.right:
                    node.right = Node(key)
                    return
                else:
                    node = node.right
            
            if node.key > key:
                if not node.left:
                    node.left = Node(key)
                    return
                else:
                    node = node.left
                    
def inorder( source ):

    if not source:
        return []

    return [*inorder(source.left), source.key, *inorder(source.right)]


def breadth(node):
    
    if not node:
        return 0
    
    l = depth(node.left)
    r = depth(node.right)
    
    return max(
        breadth(node.left), 
        l+1+r, 
        breadth(node.right)
    )


def depth(node):
    if not node:
        return 0
    return max( depth(node.left), depth(node.right) ) + 1
    
    
def get_lists( node, lists, down=0, right=0 ):
    
    if node is None:
        return
    
    lists[down][right] = f"{node.key}"
    
    get_lists( node.left,  lists, down+1, right*2 )
    get_lists( node.right, lists, down+1, right*2+1 )


def display(node):
    
    lists = [[" " for _ in range(2**l)] for l in range(depth(node))]
    get_lists(node, lists)
    
    d = len(lists) - 1
    
    for i, l in enumerate(lists):
        print( " " * ((2**d-2**i)), end="" )
        print((" ".join(l)))
    

    
bst = BST()
for i in [random.randint(0,10) for _ in range(10)]:
    bst.insert(i)

print(breadth(bst.root))
depth(bst.root)
display(bst.root)
```

```python
def counting_sort( A, K ):
    
    kimap = { k: i for i, k in enumerate(K) }
    temp = [ 0 for _ in K ]
    out = [ None for _ in A ]
    
    for a in A:
        temp[kimap[a]] += 1
    for i in range(1, len(K)):
        temp[i] += temp[i-1]
    for a in A:
        temp[kimap[a]] -= 1
        out[temp[kimap[a]]] = a
    
    return out

    
r = lambda: chr(random.randint(0,9)+97)
counting_sort([r() for _ in range(20)], [chr(x+97) for x in range(10)])
```

```python
def swap(A, i, j):
    A[i], A[j] = A[j], A[i]

    
def lomuto( A, low, high ):
    
    mid = ( low + high ) // 2
    swap(A, mid, high-1)
    
    pv = A[high-1]
    
    i = low
    for j in range(low, high):
        if A[j] < pv:
            swap(A, i, j)
            i += 1
    swap(A, i, high-1)
    
    return i


def quicksort( A, low, high ):
    
    if high - low <= 1:
        return
    
    pivot = lomuto( A, low, high )
    quicksort( A, low, pivot )
    quicksort( A, pivot+1, high )
    
    return A


def oscarsort( A ):
    
    buckets = [[] for _ in range(100)]
    
    for a in A:
        buckets[int(a)].append(a)
        
    for i in range(100):
        quicksort(buckets[i], 0, len(buckets[i]))
        
    return [ a for bucket in buckets for a in bucket ]
    
r = lambda: random.uniform(0,100)
A = [r() for _ in range(1000000)]
```

```python

def lis( A ):
    
    cache = [ 0 for _ in A ]
    
    for i, a in enumerate(A):
        cache[i] = 1
        
        for j, b in enumerate(cache[:i]):
            if A[j] < a and cache[j] >= cache[i]:
                cache[i] = cache[j] + 1
    
    return max(cache)


lis( [1,4,6,7,4,6,8,2] )
```
